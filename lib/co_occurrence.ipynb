```python
import numpy as np
import itertools
import pandas as pd
```

```python
names = []
with open('name_list.txt', 'rb') as f:
    for line in f:
        names.append(line.strip().partition("\t")[0])
f.close()
```

```python
names_v1 = []
for x in names:
    with open('sentences.txt', 'rb') as f:
        for line in f:
            if x in line or x.split(' ')[0] + ' ' in line:
                names_v1.append(x)
```

```python
allnames = sorted(list(set(names_v1)))
```

```python
for x in allnames:
    if len(x.split(' ')) == 1:
        allnames.remove(x)
    if (len(x.split(' ')) == 3 and x.split(' ')[-1].lower() == 'potter'):
        allnames.remove(x)
```

```python
names_set = []
for a,b in itertools.combinations(allnames,2):
    names_set.append([a,b])
```

```python
zero_data = np.zeros(shape=(len(allnames),len(allnames)))
family_matrix = pd.DataFrame(zero_data,index = [range(len(allnames))], columns = [range(len(allnames))])
family_matrix.index = allnames
family_matrix.columns = allnames
```

```python
import time
start_time = time.time()
#allnames = allnames[0:333]
#cooccurrence = {}
for x,y in names_set:
    with open('sentences.txt', 'rb') as f:
        for line in f:
            if (x.lower() in line.lower() or x.split(' ')[0].lower() + ' ' in line.lower() or x.split(' ')[1].lower() + ' ' in line.lower()) and (y.lower() in line.lower() or y.split(' ')[0].lower() + ' ' in line.lower() or y.split(' ')[-1].lower() + ' ' in line.lower()):
                family_matrix.loc[x,y] += 1
                family_matrix.loc[y,x] += 1
```
