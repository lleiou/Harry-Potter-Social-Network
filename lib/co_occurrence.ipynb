```python
import numpy as np
import itertools
import pandas as pd
```

```python
nick = {}
with open('name_list.txt', 'rb') as f:
    for line in f:
        if line.strip().partition("\t")[-1] != '':
            nick[line.strip().partition("\t")[0]] = line.strip().partition("\t")[0].split(' ')+line.strip().partition("\t")[-1][1:-1].split(', ')
        else:
            nick[line.strip().partition("\t")[0]] = line.strip().partition("\t")[0].split(' ')
f.close()
```

```python
names_v1 = []
for x in nick.keys():
    with open('sentences.txt', 'rb') as f:
        for line in f:
            if x.split(' ')[0] + ' ' in line:
                names_v1.append(x)
            if nick[x][2:] != []:
                for name in nick[x][2:]:
                    if name + ' ' in line:
                        names_v1.append(x)
```

```python
allnames = sorted(list(set(names_v1)))
```

```python
for x in allnames:
    if len(x.split(' ')) == 1:
        allnames.remove(x)
    if (len(x.split(' ')) == 3 and x.split(' ')[-1].lower() == 'potter'):
        allnames.remove(x)
```

```python
# need to run it again
for x in allnames:
    if len(x.split(' ')) == 1:
        allnames.remove(x)
    if (len(x.split(' ')) == 3 and x.split(' ')[-1].lower() == 'potter'):
        allnames.remove(x)
```

```python
names_set = []
for a,b in itertools.combinations(allnames,2):
    names_set.append([a,b])
```

```python
ep1 = {}
for x in allnames:
    ep1[x] = nick[x]
    
import pickle
with open("nick_ep1", "wb") as f:
    pickle.dump(ep1, f)
```

```python
zero_data = np.zeros(shape=(len(allnames),len(allnames)))
family_matrix = pd.DataFrame(zero_data,index = [range(len(allnames))], columns = [range(len(allnames))])
family_matrix.index = allnames
family_matrix.columns = allnames
```

```python
import time
start_time = time.time()
#allnames = allnames[0:333]
#cooccurrence = {}
for x,y in names_set:
    with open('sentences.txt', 'rb') as f:
        for line in f:
            for name1, name2 in itertools.product(ep1[x], ep1[y]):
                if name1.lower() + ' ' in line.lower() and name2.lower() + ' ' in line.lower():
                    family_matrix.loc[x,y] += 1
                    family_matrix.loc[y,x] += 1
```
